```python
#Importovani knihoven
from gtts import gTTS
import IPython

mytext = """Ahoj, vítám vás na další lekci Pythonu. Dnes si ukážeme zase něco nového. 
Budeme dělat for cyklus. Vypadá to, že je to skvělý nástroj. Kubo! Dneska máš volno a všechno budu učit já.
Vklidu si dej nohy nahoru a vy ostatní poslouchjte.
"""
language = 'cs'
myobj = gTTS(text=mytext, lang=language)
myobj.save("soubor.wav")
IPython.display.Audio("soubor.wav") #zobraz mi okénko
```
